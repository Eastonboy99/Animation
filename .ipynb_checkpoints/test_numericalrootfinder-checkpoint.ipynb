```python
#Start of actual fourbarlinkage approximation
#Both point 1 and for lie on the horizontal axis
from sympy import *

t1,t2,t3,t4,a,b,c,d = symbols('t1 t2 t3 t4 a b c d')
```

```python
#Vector loop equations
f1 = a*cos(t2) + b*cos(t3)-c*cos(t4) - d
f2 = a*sin(t2) + b*sin(t3)-c*sin(t4)
#Matrix equation
B = Matrix([f1,f2])
B
```

```python
#VAriables of system
var = Matrix([t3,t4])
#Jacobian matrix of B
A = B.jacobian(var)
A
```

```python
#Linkage lengths
sublinkages = {a:100,b:350,c:280,d:200}
#Initial guess of t3 and t4
xn = {t3:np.pi*45/180,t4:np.pi*60/180}
#Substitute linkage values
A = A.subs(sublinkages)
B = B.subs(sublinkages)
#Substitute initial guess values
out_t3=[]
out_t4=[]
```

```python

input_t2 = np.linspace(0,2*np.pi,100)

for j in input_t2:

    Btemp = B.subs(t2,j)
    for i in range(10):
        Atemp = A.subs(xn)
        Btemp = Btemp.subs(xn)
        dxn = Atemp.inv()*Btemp
        xn[t3] -= dxn[0]
        xn[t4] -= dxn[1]
    out_t3.append(float(xn[t3]))
    out_t4.append(float(xn[t4]))
    
```

```python
T = np.array(list(dict.values(xn)),dtype = 'float')
T = T*180/np.pi
```

```python

x4,y4 = 280*np.cos(T[1]),280*np.sin(T[1])
```
