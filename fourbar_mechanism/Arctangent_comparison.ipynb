```python
import math
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-4*math.pi,4*math.pi,10000)
y1 = np.arctan(t)

plt.plot(t,y1)

plt.show()

```

```python
#Interesting consequence of atan2, you can create a sawtooth wave from a continuous t!

y2 = np.arctan2(np.sin(t),np.cos(t))
plt.plot(t,y2)

y2 = np.arctan2(np.cos(t),np.sin(t))
plt.plot(t,y2)

plt.show()
```

```python
#Modulo'd arctangent so 0-360 degrees
y2 =(np.arctan2(np.sin(t),np.cos(t)) + 2*np.pi )% 2*np.pi
plt.plot(t,y2)

plt.show()
```
